```python
import pandas as pd
```

```python
df = pd.read_csv("ukr_migration (1).csv")
df
```

```python
filtered_df = df[df['Indicator'] == 'Migration population reduction']
filtered_df.dropna(inplace=True)
filtered_df
```

```python
average_values = filtered_df.iloc[:, 3:].mean(axis=1)
filtered_df["mean"] = average_values
```

```python
filtered_df
```

```python
df2 = pd.DataFrame()
df2["region"] = filtered_df["Region"]
df2["migration_reduction"] = filtered_df["mean"]
df2.to_csv("ukr_migration_out.csv")
```
